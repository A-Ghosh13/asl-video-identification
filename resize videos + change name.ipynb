```python
import os
from moviepy.editor import *
import shutil
```

```python
# dir = 'MS-ASLFinal8/train'
# for folder in os.listdir(dir):
    direct = 'MS-ASLFinal8/train/yes'
    for file in os.listdir(direct):
        fileDir = direct + '/{}'.format(file)
        print(fileDir)
        video = VideoFileClip(fileDir)
        vid = video.resize(height = 240)
        vid.write_videofile(file, audio_codec='aac', logger=None)   
        video.close()
        vid.close()
        shutil.move(file, fileDir)
#         print(file)
```
